```python
X = data[[ 'price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot','floors', 'waterfront', 'view', 'condition', 'sqft_above','sqft_basement', 'yr_built', 'yr_renovated']]
       
y = data["price"]
```

```python
def featurescaling(X):
    """Doing Feature scaling so that every feature came into relatively same region"""
    Xmean = X.mean(axis=0)
    X_std = X.std(axis=0)
    
    return (X - Xmean)/X_std
```

```python
X = featurescaling(X)
X.mean(axis=0),X.std(axis=0)
```

```python
# Outlier Removal 
# Removing the outlier from the dataset to get the best prediction
z_scores = (X - X.mean()) / X.std()
z_score_threshold = 3
outlier_indices = np.where(np.abs(z_scores) > z_score_threshold)
X = X.drop(outlier_indices[0])
y = y.drop(outlier_indices[0])
```

```python
X.shape,y.shape
```

```python
# Spliting the Dataset into Training and testing set
Xtrain , Xtest , ytrain , ytest = train_test_split(X,y , shuffle=False , random_state=0)
```

```python
n_features = X.shape[1]
# Adding the additional column into the X
def preprocess(X):
    if X.shape[1]==n_features:
        m = X.shape[0]
        ones = np.zeros((m,1))
        X = np.hstack((ones,X))
        
    return X

```

```python
# Defining the hypothesis function
def hypothesis(X,theta):
    """This is the hypothesis Function"""
    return np.dot(X,theta)
# Defining the costfunction
def costfunction(X,y,theta):
    """Costfunction checks how well the model is performing."""
    m = X.shape[0]
    error = np.sum((hypothesis(X,theta) - y)**2)
    return error/(2*m)

def derivative(X,y,theta):
    
    m = X.shape[0]
    y_predict = hypothesis(X,theta)
    derive = np.dot(X.T,(y_predict-y))
    
    return derive/m
```

```python
X.shape[1]
```

```python
# Defining the function to get best value of parameters
def train(X,y,alpha = 0.6 , max_itrs = 1500):
    n = X.shape[1]
    theta = np.random.randn(n)
    error_list = []
    
    for i in range(max_itrs):
        e = costfunction(X,y,theta)
        error_list.append(e)
        derive = derivative(X,y,theta)
        theta = theta - alpha*derive
        
    plt.plot(error_list)
    return theta
    
```

```python
Xtrain  = preprocess(Xtrain)
train1 = train(Xtrain , ytrain)
```

```python
Xtest = preprocess(Xtest)
y_predict = hypothesis(Xtest , train1)
```

```python
# R2score function which will give the final performance of the model 
def evaluation(y,y_predict):
    ymean = y.mean()
    num = np.sum((y - y_predict)**2)
    denom = np.sum((y - ymean)**2)
    
    return 1 - num/denom
```

```python
evaluate = evaluation(ytest,y_predict)
evaluate
```
