```python
# Dividing data into X and y Variable
X = df[['Sale', 'weight', 'resoloution', 'ppi','cpu core', 'cpu freq', 'internal mem', 'ram', 'RearCam', 'Front_Cam','battery', 'thickness']]
```

```python
y = df['Price']
```

```python
# Feature Scaling of X so to convert data into Similar Range 
def featurescaling(X):
    Xmean = X.mean(axis=0)
    Xstd = X.std(axis=0)
    
    return (X - Xmean)/Xstd
```

```python
X = featurescaling(X)
```

```python
X.head()
```

```python
X.mean(axis=0),X.std(axis=0)
```

# Selecting Best Features From the Dataset

```python
from sklearn.feature_selection import SelectKBest
```

```python
best_features = SelectKBest()
```

```python
fit = best_features.fit(X,y)
```

```python
fit.scores_
```

```python
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
featurescore = pd.concat([dfcolumns,dfscores],axis=1)
```

```python
featurescore.columns = ['Features' , 'Scores']
```

```python

featurescore
```

```python
featurescore.sort_values(by='Scores' , ascending = False)
```

```python
X_new = X[['battery','weight', 'RearCam' , 'resoloution' , 'internal mem' , 'ppi' , 'ram', 'Front_Cam',  'thickness' , 'cpu core' , 'cpu freq']]
```

```python
X_new.head()
```

# Applying Machine Learning Linear Regression Algorithm

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
Xtrain , Xtest , ytrain , ytest = train_test_split(X_new , y , test_size=0.3)
```

```python
model = LinearRegression()
```

```python
model.fit(Xtrain , ytrain)
```

```python
y_predict = model.predict(Xtest)
```

# Model evaluation

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(ytest,y_predict)
```
